```python
# 1. Removendo as linhas duplicadas
df = df.drop_duplicates()

# 2. Resetando o index (para a contagem das linhas ficar certinha de 0 até o final)
df = df.reset_index(drop=True)

# 3. Verificando o novo tamanho da base
print(f"Duplicatas removidas! Novo total de registros: {len(df)}")
```

```python
#Após remoção das duplicadas
df.duplicated().sum()
```

```python
# 2. Lista das variáveis float que você deseja transformar
colunas_para_inteiro = ['Age', 'FCVC', 'NCP', 'CH2O', 'FAF', 'TUE']

# 3. Aplicando o arredondamento para o mais próximo (.round())
# e convertendo para inteiro (.astype(int))
df[colunas_para_inteiro] = df[colunas_para_inteiro].round().astype(int)
```

```python
df.head(2)
```

```python
df.head()
```

```python
# 1. Criando o dicionário de mapeamento para todas as binárias
# Mapeamos Gênero e as respostas Sim/Não simultaneamente
mapeamento_binario = {
    'Female': 0, 'Male': 1,
    'no': 0, 'yes': 1
}

# 2. Lista das colunas que queremos transformar
colunas_binarias = ['Gender', 'family_history', 'FAVC', 'SMOKE', 'SCC']

# 3. Aplicando a transformação com .replace()
# O replace mantém o valor original se não encontrar a chave, evitando NaNs em execuções duplas
df[colunas_binarias] = df[colunas_binarias].replace(mapeamento_binario)

```

```python
df.head()
```

```python
#para analisar uma coluna específica e saber quais informações há nela
set(df["CAEC"])
```

```python
# Definindo a hierarquia lógica
mapeamento_caec = {
    'no': 0,
    'Sometimes': 1,
    'Frequently': 2,
    'Always': 3
}

# Aplicando de forma segura com .replace()
df['CAEC'] = df['CAEC'].replace(mapeamento_caec)

# Verificando a transformação
print(df['CAEC'].value_counts())
```

```python
df.head()
```

```python
#para analisar uma coluna específica e saber quais informações há nela
set(df["CALC"])
```

```python
# 1. Definindo o mapeamento para CALC (seguindo a mesma escala da CAEC)
mapeamento_calc = {
    'no': 0,
    'Sometimes': 1,
    'Frequently': 2,
    'Always': 3
}

# 2. Aplicando a transformação com .replace() para evitar erros de execução dupla
df['CALC'] = df['CALC'].replace(mapeamento_calc)

# 3. Verificando se a conversão deu certo
print("Valores únicos de CALC após conversão:")
print(df['CALC'].unique())
```

```python
df.head()
```

```python
#analisando quantidade pessoas que bebem e frequencia
print(df['CALC'].value_counts())
```

```python
#Agora para tratar a coluna MTRANS precisamos transformar os valore sem colunas e preencher com 1 para quem tiver naquela opção
```

```python
# Criando colunas separadas para cada meio de transporte
# O parâmetro 'dtype=int' garante que o resultado seja 0 e 1 (e não True/False)
df = pd.get_dummies(df, columns=['MTRANS'], prefix='transporte', dtype=int)

# Verificando as novas colunas criadas
print(df.columns)
print(df.filter(like='transporte').head())
```

```python
df.head()
```

```python
#analisando a variável target
print(df['Obesity'].value_counts())
```

```python
# 1. Definindo a ordem lógica (do menor peso para o maior nível de obesidade)
mapeamento_obesity = {
    'Insufficient_Weight': 0,
    'Normal_Weight': 1,
    'Overweight_Level_I': 2,
    'Overweight_Level_II': 3,
    'Obesity_Type_I': 4,
    'Obesity_Type_II': 5,
    'Obesity_Type_III': 6
}

# 2. Aplicando a transformação de forma segura
df['Obesity'] = df['Obesity'].replace(mapeamento_obesity)

# 3. Verificando o resultado
print("Níveis transformados:")
print(df['Obesity'].value_counts().sort_index())
```

```python
#Estudar a correlação entre as variáveis
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Calculando a matriz de correlação
corr_matrix = df.corr().round(2)

# 2. Criando o gráfico de calor (Heatmap)
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

plt.title('Mapa de Calor: Correlação entre Hábitos e Nível de Obesidade', fontsize=16)
plt.tight_layout()
plt.show()

# 3. Extraindo apenas a correlação com a variável alvo para facilitar a leitura
print("Correlação direta com o Nível de Obesidade:")
print(corr_matrix['Obesity'].sort_values(ascending=False))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Separando os dados
# X = todas as colunas exceto a nossa 'Target' (Obesity)
# y = apenas a coluna 'Obesity'
X = df.drop(columns=['Obesity'])
y = df['Obesity']

# 2. Divisão entre Treino (80%) e Teste (20%)
# O random_state garante que o resultado seja replicável
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

# 3. Criando e Treinando o Modelo
# Usaremos 100 árvores de decisão para compor nossa floresta
modelo_final = RandomForestClassifier(n_estimators=100, random_state=42)
modelo_final.fit(X_train, y_train)

# 4. Fazendo as Previsões com os dados que o modelo nunca viu (X_test)
y_pred = modelo_final.predict(X_test)

# 5. Avaliando o Desempenho
acuracia = accuracy_score(y_test, y_pred)
print(f"🚀 Acurácia Final do Modelo: {acuracia:.2%}")
print("\n📋 Relatório de Classificação:\n")
print(classification_report(y_test, y_pred))

# 6. Visualizando a Matriz de Confusão (Para ver onde o modelo erra)
plt.figure(figsize=(10, 8))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues')
plt.title('Matriz de Confusão: Previsão de Obesidade')
plt.xlabel('Previsão do Modelo')
plt.ylabel('Valor Real (Hospital)')
plt.show()
```

```python
# Criando o gráfico de Importância das Variáveis
importances = modelo_final.feature_importances_
features = X.columns
df_importances = pd.DataFrame({'feature': features, 'importance': importances}).sort_values(by='importance', ascending=False)

plt.figure(figsize=(8, 6))
sns.barplot(x='importance', y='feature', data=df_importances, palette='magma')
plt.title('Importância das Variáveis no Modelo (Random Forest)', fontsize=14)
plt.xlabel('Nível de Importância')
plt.ylabel('Variáveis')
plt.show()
```

```python
from sklearn.metrics import classification_report
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Gerando o relatório como um dicionário
report = classification_report(y_test, y_pred, output_dict=True)

# 2. Convertendo para DataFrame e limpando para o gráfico
# Removemos 'accuracy', 'macro avg' e 'weighted avg' para focar nas classes
df_report = pd.DataFrame(report).transpose().iloc[:7, :]

# 3. Criando o gráfico de barras para o F1-Score
plt.figure(figsize=(12, 6))
ax = sns.barplot(x=df_report.index, y=df_report['f1-score'], palette='summer')

# 4. Customização profissional
plt.title('F1-Score por Nível de Obesidade (Avaliação de Equilíbrio)', fontsize=15)
plt.xlabel('Níveis de Obesidade', fontsize=12)
plt.ylabel('F1-Score (0 a 1)', fontsize=12)
plt.ylim(0, 1.1) # Deixar um espaço para os números

# Adicionando os nomes reais no eixo X
niveis_nomes = ['Baixo Peso', 'Normal', 'Sobrepeso I', 'Sobrepeso II', 'Obesity I', 'Obesity II', 'Obesity III']
plt.xticks(ticks=range(7), labels=niveis_nomes, rotation=15)

# Adicionando os valores em cima das barras
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.2f'),
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha = 'center', va = 'center', xytext = (0, 9), textcoords = 'offset points', fontweight='bold')

plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.show()
```

```python
#identifiquei uma varíavel obvia que foge do objetivo de dar visibilidade para o médico sobre o paciente pode ou não ser obeso no futo, a variável peso atrapalha meu modelo atual, então vou rodar sem essa variável.
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

# 1. Separando os dados REMOVENDO as variáveis óbvias
# Queremos prever a obesidade APENAS por comportamento e genética
X_preventivo = df.drop(columns=['Obesity', 'Weight', 'Height'])
y = df['Obesity']

# 2. Divisão Treino/Teste
X_train, X_test, y_train, y_test = train_test_split(X_preventivo, y, test_size=0.2, random_state=42)

# 3. Treinando o modelo "Médico Preventivo"
modelo_preventivo = RandomForestClassifier(n_estimators=100, random_state=42)
modelo_preventivo.fit(X_train, y_train)

# 4. Resultados
y_pred = modelo_preventivo.predict(X_test)
acuracia_real = accuracy_score(y_test, y_pred)

print(f"🚀 Acurácia do Modelo Preventivo (Sem Peso/Altura): {acuracia_real:.2%}")
print("\n📋 Relatório de Classificação:\n", classification_report(y_test, y_pred))
```

```python

import joblib
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# 1. Definição da Semente (Reprodutibilidade)
SEED = 42

# 2. Carregamento e Preparação dos Dados
# (Assumindo que seu df já passou pelas transformações numéricas que fizemos)
X = df.drop(columns=['Obesity', 'Weight', 'Height']) # Foco em Prevenção
y = df['Obesity']

# 3. Divisão de Treino e Teste
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=SEED, stratify=y
)

# 4. Treinamento do Modelo
# Usamos stratify=y para garantir que a proporção de obesos seja igual no treino e teste
modelo_preventivo = RandomForestClassifier(n_estimators=100, random_state=SEED)
modelo_preventivo.fit(X_train, y_train)

# 5. Validação Rápida
y_pred = modelo_preventivo.predict(X_test)
print(f"✅ Modelo treinado com sucesso! Acurácia: {accuracy_score(y_test, y_pred):.2%}")

# 6. EXPORTAÇÃO (O passo crucial para o Streamlit)
# Vamos salvar o modelo e a lista de colunas para garantir que o Deploy não quebre
joblib.dump(modelo_preventivo, 'modelo_obesidade.pkl')
joblib.dump(X.columns.tolist(), 'colunas_modelo.pkl')

print("💾 Arquivos 'modelo_obesidade.pkl' e 'colunas_modelo.pkl' salvos!")
```
